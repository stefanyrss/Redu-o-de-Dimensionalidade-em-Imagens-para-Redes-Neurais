```python
from google.colab import files
uploaded = files.upload()

# Verificando os arquivos carregados
import io
import cv2
import numpy as np
from PIL import Image

for fn in uploaded.keys():
    print('User uploaded file: "{name}"'.format(name=fn))

    # Lendo a imagem diretamente de bytes
    img = Image.open(io.BytesIO(uploaded[fn]))
    img.show()

    # Convertendo para um formato adequado para o OpenCV (se necessário)
    img_cv = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
    cv2.imwrite('/content/lena.png', img_cv)  # Salvando a imagem convertida
```

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem colorida, corrigindo o caminho do arquivo:
image = cv2.imread('/content/lena.png')  # Alterado para 'lena.png', que foi salvo na célula anterior.

# Verificar se a imagem foi carregada corretamente
if image is None:
    print("Erro ao carregar a imagem. Verifique o caminho do arquivo.")
else:
    # Converter para imagem em escala de cinza (0 a 255)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Converter para imagem binarizada (0 e 255)
    _, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

    # Exibir as imagens
    plt.figure(figsize=(12, 4))

    # Exibir imagem original colorida
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Imagem Colorida')
    plt.axis('off')

    # Exibir imagem em níveis de cinza
    plt.subplot(1, 3, 2)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Imagem em Níveis de Cinza')
    plt.axis('off')

    # Exibir imagem binarizada
    plt.subplot(1, 3, 3)
    plt.imshow(binary_image, cmap='gray')
    plt.title('Imagem Binarizada')
    plt.axis('off')

    plt.show()
```
